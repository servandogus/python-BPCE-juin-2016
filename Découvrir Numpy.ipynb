```python
#Générer un array avec des valeurs entre 0 et 8, 
#afficher les éléments 1 à 6 puis modifiez les éléments 6 à 8 
#en leur fixant la valeur de 22.
arr3=np.arange(9)
arr3[1:7]
arr3[6:9]=22
arr3
```

```python
#Construire une matrice rectangulaire de taille 3 x 2 avec une suite de nombres 
#commençant à 5, calculer le produit de la transposée par la matrice.
arr4=np.arange(5,11).reshape(3,2)
arr5=np.dot(arr4.T,arr4)
arr5
```
